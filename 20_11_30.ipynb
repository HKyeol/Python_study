# 29. 시퀀스 자료 연결 이해하기(+)

문자열 + 문자열, 리스트 + 리스트, 튜플 + 튜플과 같이 두개의 동일한 시퀀스 자료형에 대해 '+' 연산자로 연결이 가능합니다.


예제 :

````
a = 'I love '
b = 'python'
print( a + b)
````

오라클은 ||인데 파이썬은 + 이다.
아래의 SQL을 판다스로 구현하시오!

SQL> select ename || sal
  from emp

  

````
for i, k in zip(emp['ename'], emp['sal']):
  print (i + str(k))
````

설명: zip 함수를 for loop문에 사용하게 되면 두 개의 범위 데이터 한번에 받아서 loop를 수행할 수 있습니다. str 함수를 사용한 이유는 문자형 + 문자형을 연결이 가능한데 문자형 + 숫자형은 연결이 안됩니다. 그래서 숫자형을 문다형으로 변환하기 위해서 str 함수를 사용했습니다.

문제121.(통계를 코드로 구현 문제 4번의 링크2)  
어느 비행기 탑승객의 짐의 무게는 평균이 18kg이고 표준편차가 3kg인 정규분포를 따른다고 한다. 이 비행기 탑승객 중에서 36명을 임의추출할 대, 짐의 평균 무게가 17kg 이상일 확률을 구하여라. (단, Z가 표준정규분포를 따르는 확률변수일 때, P(0<= Z <= 2) = 0.4772이다)

````
import numpy as np
# 짐의 무게 N ~ (18, 3^2)

n = 36
mu = 18
std = 3/np.sqrt(n)


Z = (17 - mu)/std   # Z = -2이고  P(0<= Z <= 2) = 0.4772 이므로
P = 0.5 + 0.4772
print(P)
````
